```python
def makePgNum(num):
    if num == 1:
        return num
    elif num == 0:
        return num + 1
    else:
        return num+9*(num-1)
```

```python
def news_contents_crawler(news_url):
    contents = []
    for i in news_url:
        news = requests.get(i)
        news_html = BeautifulSoup(news.text, "html.parser")
        contents.append(news_html.find_all('p'))
    return contents
```
