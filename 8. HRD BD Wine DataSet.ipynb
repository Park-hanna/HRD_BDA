# 검정

```python
from scipy import  stats
from statsmodels.formula.api import ols, glm
```

```python
red_wine_quality = wine.loc[wine['type']=='red','quality']
```

```python
white_wine_quality = wine.loc[wine['type'] == 'white','quality']
```

ues scipy package stats.ttest_ind() function and t-test, difference between the two groups 

```python
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
```

linear regression equation <br>: dependent variable -> quality / independent variable -> excluding type and quality <br>
Rformula = 'quality ~ fixed_acidity + volatile_acidity_ citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'

```python
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
```

```python
regression_result = ols(Rformula, data=wine).fit()
```

```python
regression_result.summary()
```

# 부분 회귀 플롯으로 시각화

```python
import statsmodels.api as sm
```

fixed_acidity가 종속 변수 quality에 미치는 영향력 시각화 <br>
부분 회귀에 사용한 독립 변수와 종속 변수를 제외한 나머지 변수 이름을 list others로 추출

```python
others = list(set(wine.columns).difference(set(['quality','fixed_acidity'])))
```

나머지 변수 고정, fixed_acidity가 종속 변수 quality에 미치는 영향에 부분 회귀 수행 <br>
부분 회귀의 결과가 플롯으로 시각화되어 나타남

```python
p, resides = sm.graphics.plot_partregress('quality','fixed_acidity', others, data = wine, ret_coords = True)
```
