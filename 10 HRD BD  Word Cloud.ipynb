```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
```

```python
with open('C:/Users/ParkHanna/Desktop/HRD/BigData/data/대한민국헌법.txt','rt',encoding='utf-8') as f:
    data = f.read()

okt=Okt()
nouns = okt.nouns(data)
words = [i for i in nouns if len(i)>1]
c= Counter(words)
```

# 한글 워드 클라우드 만들기

워드 클라우드 생성을 위해 사용할 큰 템플릿 준비 <br>
템플릿 정보 : 그림의 크기 (width, height, scale, max_font_size) <br>
제일 중요한 것 : 한글 폰트 지정 <br>
윈도우 사용자 기준 폰트 정보 : 'C:\Windows\Fonts' <br>
해당 폴더에서 폰트 확인후 원하는 폰트 파일 속성에서 폰트의 파일명 복사 후 워드 클라우드에 사용할 폰트를 wordCloud 객체를 생성할때 font_path 파라미터에 폰트 파일명을 넣어줌 (폰트는 한글을 지원해야 사용가능)

```python
wc = WordCloud(font_path='YES24', width=100, height=60, scale=2.0,max_font_size=250) 
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
```

```python
img = Image.open('c:/Users/ParkHanna/Desktop/HRD/BigData/data/apple.jpg')
img_array = np.array(img)

wc = WordCloud(font_path="YES24", width=100, height=60, scale=2.0, max_font_size =250, mask=img_array)
gen = wc.generate_from_frequencies(c)

plt.figure()
plt.imshow(gen)
```
