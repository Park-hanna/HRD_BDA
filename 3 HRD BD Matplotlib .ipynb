```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [100, 200, 300, 400]
y = [10, 30, 50, 70]

plt.plot(x,y)
plt.title("1st Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.savefig('C:/Users/ParkHanna/HRD BD/1stplotpic.png')
plt.show()
```

```python
x = np.linspace(0, np.pi * 30, 1000)

fig, axes = plt.subplots(2, 1)

axes[0].plot(x, np.sin(x))

axes[1].plot(x, np.cos(x))

fig.savefig('C:/Users/ParkHanna/HRD BD/2ndplotpic.png')
```

```python
x = np.arange(-9, 10)
y = x ** 2

plt.plot(x,y, linestyle =":", marker="*")
plt.show()
```

```python
y1 = x ** 2
y2 = -x

plt.plot(x, y1, linestyle="-", marker="*", color="red", label="y = x * x")
plt.plot(x, y2, linestyle=":", marker="o", color="blue", label="y = -x")
plt.xlabel("X")
plt.ylabel("Y")

plt.legend(
    shadow = True,
    borderpad = 1
)
plt.show()
```

```python
plt.plot(
    x, y1,
    linestyle = ":",
    marker = "o",
    markersize = 8,
    markerfacecolor = "blue",
    markeredgecolor = "black")
plt.show()
```

```python
plt.bar(x, x**2)
plt.show()
```

```python
x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)

data = [x, y ,z]

x_array = np.arange(10)

for i in range(0, 3):
    plt.bar(
        x_array,
        data[i],
        bottom = np.sum(data[:i], axis=0)
    )
    
plt.show()
```
